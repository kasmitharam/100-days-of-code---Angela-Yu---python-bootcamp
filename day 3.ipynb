```python
#Exercise 2
"""Instructions
Write a program that interprets the Body Mass Index (BMI) based on a user's weight and height.

It should tell them the interpretation of their BMI based on the BMI value.

Under 18.5 they are underweight
Over 18.5 but below 25 they have a normal weight
Over 25 but below 30 they are slightly overweight
Over 30 but below 35 they are obese
Above 35 they are clinically obese."""


# ðŸš¨ Don't change the code below ðŸ‘‡
height = float(input("enter your height in m: "))
weight = float(input("enter your weight in kg: "))
# ðŸš¨ Don't change the code above ðŸ‘†

#Write your code below this line ðŸ‘‡
BMI = round((weight/(height * height)))

if BMI < 18.5:
    print(f"Your BMI is {BMI}, you are underweight.")
elif BMI > 18.5 and BMI < 25:
    print(f"Your BMI is {BMI}, you have a normal weight.")
elif BMI > 25 and BMI < 30:
    print(f"Your BMI is {BMI}, you are slightly overweight.")
elif BMI > 30 and BMI < 35:
    print(f"Your BMI is {BMI}, you are obese.")

else:
    print(f"Your BMI is {BMI}, you are clinically obese.")
```

```python
#exercise 3
"""Instructions
Write a program that works out whether if a given year is a leap year. A normal year has 365 days, leap years have 366, with an extra day in February. The reason why we have leap years is really fascinating, this video does it more justice:

https://www.youtube.com/watch?v=xX96xng7sAE

This is how you work out whether if a particular year is a leap year.

on every year that is evenly divisible by 4 

**except** every year that is evenly divisible by 100 

**unless** the year is also evenly divisible by 400

e.g. The year 2000:

2000 Ã· 4 = 500 (Leap)

2000 Ã· 100 = 20 (Not Leap)

2000 Ã· 400 = 5 (Leap!)

So the year 2000 is a leap year.

But the year 2100 is not a leap year because:

2100 Ã· 4 = 525 (Leap)

2100 Ã· 100 = 21 (Not Leap)

2100 Ã· 400 = 5.25 (Not Leap)

Warning your output should match the Example Output format exactly, even the positions of the commas and full stops."""
# ðŸš¨ Don't change the code below ðŸ‘‡
year = int(input("Which year do you want to check? "))
# ðŸš¨ Don't change the code above ðŸ‘†

#Write your code below this line ðŸ‘‡
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap year.")
        else:
            print("Not leap year.")
    else:
        print("Leap Year.")
else:
    print("Not leap year.")
```

```python
#Exercise 4
"""
Instructions
Congratulations, you've got a job at Python Pizza. Your first job is to build an automatic pizza order program.

Based on a user's order, work out their final bill.

Small Pizza: $15

Medium Pizza: $20

Large Pizza: $25

Pepperoni for Small Pizza: +$2

Pepperoni for Medium or Large Pizza: +$3

Extra cheese for any size pizza: + $1
"""

# ðŸš¨ Don't change the code below ðŸ‘‡
print("Welcome to Python Pizza Deliveries!")
size = input("What size pizza do you want? S, M, or L ")
add_pepperoni = input("Do you want pepperoni? Y or N ")
extra_cheese = (input("Do you want extra cheese? Y or N ")).strip()
print(type(extra_cheese))
# ðŸš¨ Don't change the code above ðŸ‘†

#Write your code below this line ðŸ‘‡
small = 15
medium = 20
large = 25

pep_small = 2
pep_med_lar = 3

total_bill_amount = 0

if size == 'S':
    total_bill_amount = total_bill_amount + small
    if add_pepperoni == 'Y':
        total_bill_amount = total_bill_amount +  pep_small

elif size == 'M':
    total_bill_amount = total_bill_amount + medium
    if add_pepperoni == 'Y':
        total_bill_amount = total_bill_amount +  pep_med_lar

elif size == 'L':
    total_bill_amount = total_bill_amount + large
    if add_pepperoni == 'Y':
        total_bill_amount = total_bill_amount +  pep_med_lar

if 'Y' in extra_cheese:
    print("yesssssss")
    total_bill_amount = total_bill_amount + 1
else:
    print("NOOOO")

print(f"Your final bill is: {total_bill_amount}.")



```

```python
#Exercise 5
"""Instructions
You are going to write a program that tests the compatibility between two people.

To work out the love score between two people:

Take both people's names and check for the number of times the letters in the word TRUE occurs. 

Then check for the number of times the letters in the word LOVE occurs. 

Then combine these numbers to make a 2 digit number.

For Love Scores less than 10 or greater than 90, the message should be:"""



# ðŸš¨ Don't change the code below ðŸ‘‡
print("Welcome to the Love Calculator!")
name1 = ((input("What is your name? \n")).replace(" ","")).lower()
name2 = ((input("What is their name? \n")).replace(" ","")).lower()
# ðŸš¨ Don't change the code above ðŸ‘†

#Write your code below this line ðŸ‘‡
both_name = name1+name2

love_score = 0
true_score = 0

if 't' in both_name:
    true_score += both_name.count('t')
if 'r' in both_name:
    true_score += both_name.count('r')
if 'u' in both_name:
    true_score += both_name.count('u')
if 'e' in both_name:
    true_score += both_name.count('e')
if 'l' in both_name:
    love_score += both_name.count('l')
if 'o' in both_name:
    love_score += both_name.count('o')
if 'v' in both_name:
    love_score += both_name.count('v')
if 'e' in both_name:
    love_score += both_name.count('e')

true_love = int(str(true_score)+str(love_score))

if true_love < 10 or true_love > 90:
    print(f"Your score is {true_love}, you go together like coke and mentos.")
elif true_love >= 40 and true_love <= 50:
    print(f"Your score is {true_love}, you are alright together.")
else:
    print(f"Your score is {true_love}.")
```
